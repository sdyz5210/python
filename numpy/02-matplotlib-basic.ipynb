# Matplotlib基础

Matplotlib 可能是 Python 2D-绘图领域使用最广泛的套件。它能让使用者很轻松地将数据图形化，并且提供多样化的输出格式。
导入 matplotlib 和 numpy：

```python
%pylab
```

## plot 二维图

用法：  
plot(y)  
plot(x,y)  
plot(x,y,format_string)  

只给定 y 值，默认以下标为 x 轴：

```python
%matplotlib inline
x = linspace(0,2*pi,50)
y =  sin(x)
plot(y)
```

给定x、y的值

```python
plot(x,y)
```

多条曲线的情况

```python
plot(x,sin(x),x,sin(2*x))
```

使用字符串，给定线条参数：

```python
plot(x,sin(x),"r-^")
```

```python
plot(x, sin(x), 'b-o',
    x, sin(2 * x), 'r-^')
```

## scatter 散点图

scatter(x, y)  
scatter(x, y, size)  
scatter(x, y, size, color)  

```python
plot(x,sin(x),'bo')
```

可以使用 scatter 达到同样的效果

```python
scatter(x,sin(x))
```

scatter函数与Matlab的用法相同，还可以指定它的大小、颜色等参数

```python
x = rand(200)
y = rand(200)
size = rand(200)*30
color = rand(200)
scatter(x,y,size,color)
# 显示颜色条
colorbar()
```

## 多图
使用figure()命令产生新的图像：

```python
t = linspace(0, 2*pi , 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```

也可以使用subplot在一副画中画多副子图：  
subplot(row,colum,index)

```python
subplot(1,2,1)
plot(x)

subplot(1,2,2)
plot(y)
```

## 向图中添加数据

默认多次plot会叠加

```python
plot(x)
plot(y)
```

也可以使用Matlab类似用hold(Flase)关掉，这样新图会覆盖原图

```python
plot(x)
hold()
plot(y)
# hold(True)
```

## 标签

可以在plot中加入label，使用legend加上图例

```python
plot(x,label="sin")
plot(y,label="cos")
legend()
```

或者直接在legend中加入：

```python
plot(x)
plot(y)
legend(['sin','cos'])
```

## 坐标轴、标题、风格

可以设置坐标轴的标签和标题：

```python
plot(x,sin(x))
xlabel('radians')
ylabel('amplitude',fontsize='large')
title('Sin(x)')
```

用grid()来显示网络：

```python
plot(x,sin(x))
xlabel('radians')
ylabel('amplitude',fontsize='large')
title("sin(x)")
grid()
```

## 清除、关闭图像

清除已有的图像使用： 

    clf()

关闭当前图像：  

    close()
    
关闭所有图像：

    close('all')
