```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#read_table可以读取文本文件，默认\t分割

df = pd.read_table('T17TP61540.rcp', header=0)
df.head(5)
```

```python
df['new_pos'] = df['pos']/1000000
df1 = df[df['chr']=='chr1']
ax = df1.plot.scatter(x='new_pos',y='rcp', c="#7f7f7f", s=6. ,figsize=(16,2.5))
ax.set_xlabel('chromosome pos(Mb)')
ax.set_ylabel('Copy Number')
```
